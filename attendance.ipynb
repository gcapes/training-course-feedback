# Attendance
A bit of a look at which faculties attended which courses

```python
import pandas as pd
```

```python
colHeaders = ['timeStamp','fullName','email','faculty','promo','languages','vcs','softEng','course','rating']
colsToLoad = list(range(0,9))
colsToLoad.append(13)
data = pd.read_csv('/home/mbexegc2/Documents/training/training-course-feedback/'
                   'data/feedback.tsv', delimiter='\t', usecols=colsToLoad, names=colHeaders, skiprows=1)
```

## Which faculty attended the most courses?
FSE attended more courses than the other departments combined.

```python
data.groupby('faculty').count()
```

## Which courses were most popular?
This varies by faculty. R, Python, Git, and LaTeX

```python
data.groupby(['faculty', 'course']).count()
```

```python
data.groupby('faculty')['course'].value_counts()
```

```python
data.groupby('faculty')['course'].value_counts().plot(kind='barh')
```
